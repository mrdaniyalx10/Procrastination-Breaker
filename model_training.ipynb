```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
import pickle
import warnings
warnings.filterwarnings('ignore')

print("Block 1: Libraries loaded successfully, Bro!")
```

# Procrastination-Breaker: AI Brain Training
This notebook trains the machine learning models that power the Procrastination-Breaker system.

```python
# Simulated phone usage log (Hour of day, App, and Minutes spent)
data = {
    'Hour': [8, 9, 13, 14, 22, 23, 1, 2, 10, 15],
    'App_Name': ['WhatsApp', 'Chrome', 'Instagram', 'TikTok', 'TikTok', 'PUBG', 'Instagram', 'TikTok', 'Docs', 'YouTube'],
    'Duration_Minutes': [5, 10, 30, 45, 60, 120, 40, 50, 15, 25]
}

df = pd.DataFrame(data)

# Convert App Names to Numbers (AI only reads numbers)
encoder = LabelEncoder()
df['App_Code'] = encoder.fit_transform(df['App_Name'])

print("Block 2: Data loaded! Here is a peek:")
print(df.head())
```

```python
# 1. K-MEANS CLUSTERING (Groups your habits into 3 Risk Zones)
X_clustering = df[['Hour', 'App_Code']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clustering)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Risk_Cluster'] = kmeans.fit_predict(X_scaled)

# Assign Risk Labels to the clusters
df['Risk_Label'] = df['Risk_Cluster'].map({0: 'Low', 1: 'Medium', 2: 'High'})

# 2. DECISION TREE (Learns to predict the risk instantly)
X_tree = df[['Hour', 'App_Code']]
y_tree = df['Risk_Label']

tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_tree, y_tree)

print("Block 3: K-Means and Decision Tree Models Trained Successfully!")
```

```python
# Simulate the AI learning which intervention stops you from scrolling
states = [0, 1, 2] # The 3 risk clusters
actions = ['Math Puzzle', 'Deep Breathing', '30s Screen Lock']

# Initialize an empty Q-Table
q_table = np.zeros((len(states), len(actions)))

# Simulate the rewards (Pretending the AI learned what works best for you)
q_table[2, 0] = 9.5  # Math Puzzle works best for Cluster 2 (Late Night)
q_table[1, 2] = 8.0  # Screen Lock works best for Cluster 1 (Afternoon)
q_table[0, 1] = 5.0  # Breathing works best for Cluster 0 (Morning)

print("Block 4: Q-Learning Table Initialized!")
print(pd.DataFrame(q_table, columns=actions, index=['Cluster 0', 'Cluster 1', 'Cluster 2']))
```

```python
# Pack all our models into a dictionary and save it to your laptop
project_brain = {
    'encoder': encoder,
    'scaler': scaler,
    'kmeans': kmeans,
    'tree_model': tree_model,
    'q_table': q_table,
    'actions': actions
}

with open('model.pkl', 'wb') as file:
    pickle.dump(project_brain, file)

print("Block 5: SUCCESS! model.pkl has been saved to your folder.")
```
